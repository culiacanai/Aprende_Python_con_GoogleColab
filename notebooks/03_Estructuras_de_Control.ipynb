## 1. Condicionales: if, elif, else

Los condicionales permiten que tu programa tome decisiones. La estructura es:

```python
if condici√≥n:
    # se ejecuta si la condici√≥n es True
elif otra_condici√≥n:
    # se ejecuta si la primera fue False y esta es True
else:
    # se ejecuta si todas las anteriores fueron False
```

> ‚ö†Ô∏è **Importante:** En Python la **indentaci√≥n** (4 espacios) define qu√© c√≥digo pertenece a cada bloque. No es decoraci√≥n, es sintaxis.

### 1.1 if simple


````python
temperatura = 40

if temperatura > 35:
    print("ü•µ ¬°Hace mucho calor en Culiac√°n hoy!")
    print("Lleva agua y protector solar")
````

### 1.2 if / else


````python
edad = int(input("¬øCu√°ntos a√±os tienes? "))

if edad >= 18:
    print("‚úÖ Eres mayor de edad")
else:
    print(f"‚ùå Eres menor de edad, te faltan {18 - edad} a√±os")
````

### 1.3 if / elif / else

Cuando hay m√°s de dos opciones:


````python
# Sistema de calificaciones
calificacion = float(input("¬øCu√°l fue tu calificaci√≥n? (0-100): "))

if calificacion >= 90:
    letra = "A"
    mensaje = "¬°Excelente! üåü"
elif calificacion >= 80:
    letra = "B"
    mensaje = "Muy bien üëè"
elif calificacion >= 70:
    letra = "C"
    mensaje = "Bien, pero puedes mejorar üí™"
elif calificacion >= 60:
    letra = "D"
    mensaje = "Necesitas estudiar m√°s üìö"
else:
    letra = "F"
    mensaje = "Reprobado üòü"

print(f"Calificaci√≥n: {calificacion} ‚Üí {letra}")
print(mensaje)
````

### 1.4 Condicionales anidados

Puedes poner un `if` dentro de otro `if`:


````python
tiene_cita = True
tiene_credencial = True
edad = 25

if tiene_cita:
    if tiene_credencial:
        if edad >= 18:
            print("‚úÖ Puede pasar a consulta")
        else:
            print("‚ö†Ô∏è Necesita ir con un adulto")
    else:
        print("‚ùå Necesita traer su credencial")
else:
    print("‚ùå Necesita agendar una cita primero")
````

> üí° **Tip:** Si tienes muchos niveles anidados, a veces es mejor usar `and` para simplificar:


````python
# Versi√≥n simplificada del ejemplo anterior
if tiene_cita and tiene_credencial and edad >= 18:
    print("‚úÖ Puede pasar a consulta")
elif not tiene_cita:
    print("‚ùå Necesita agendar una cita primero")
elif not tiene_credencial:
    print("‚ùå Necesita traer su credencial")
else:
    print("‚ö†Ô∏è Necesita ir con un adulto")
````

### 1.5 Operador ternario (if en una l√≠nea)

Para condiciones simples, Python permite escribir el if en una sola l√≠nea:


````python
edad = 25

# Forma larga
if edad >= 18:
    status = "adulto"
else:
    status = "menor"

# Forma corta (ternario)
status = "adulto" if edad >= 18 else "menor"
print(status)

# Otro ejemplo
temperatura = 38
clima = "calor ü•µ" if temperatura > 30 else "fresco üòé"
print(f"Hace {clima}")
````

## 2. Ciclo for: repetir con una lista

El ciclo `for` recorre cada elemento de una secuencia (lista, string, rango, etc.) y ejecuta un bloque de c√≥digo para cada uno.

### 2.1 Recorrer una lista


````python
# Recorrer una lista de ciudades
ciudades = ["Culiac√°n", "Mazatl√°n", "Los Mochis", "Guasave", "Navolato"]

for ciudad in ciudades:
    print(f"üìç {ciudad}, Sinaloa")
````

````python
# Recorrer una lista de n√∫meros
ventas_semanales = [15000, 22000, 18500, 25000, 19000]

total = 0
for venta in ventas_semanales:
    total += venta
    print(f"Venta: ${venta:,} ‚Äî Acumulado: ${total:,}")

print(f"
Total de la semana: ${total:,}")
print(f"Promedio diario: ${total / len(ventas_semanales):,.0f}")
````

### 2.2 Recorrer un string

Un string es una secuencia de caracteres, as√≠ que tambi√©n se puede recorrer:


````python
palabra = "PYTHON"

for letra in palabra:
    print(f"üî§ {letra}")
````

### 2.3 range(): generar secuencias de n√∫meros

`range()` genera una secuencia de n√∫meros sin crear una lista en memoria:

| Forma | Qu√© genera | Ejemplo |
|-------|-----------|---------|
| `range(n)` | 0 hasta n-1 | `range(5)` ‚Üí 0, 1, 2, 3, 4 |
| `range(inicio, fin)` | inicio hasta fin-1 | `range(1, 6)` ‚Üí 1, 2, 3, 4, 5 |
| `range(inicio, fin, paso)` | Con saltos | `range(0, 10, 2)` ‚Üí 0, 2, 4, 6, 8 |


````python
# Contar del 1 al 10
for i in range(1, 11):
    print(i, end=" ")  # end=" " evita el salto de l√≠nea
print()  # Salto de l√≠nea al final

# Tabla del 7
print("
Tabla del 7:")
for i in range(1, 11):
    print(f"7 √ó {i} = {7 * i}")
````

````python
# Cuenta regresiva
print("üöÄ Cuenta regresiva:")
for i in range(10, 0, -1):
    print(f"  {i}...")
print("  ¬°DESPEGUE! üéâ")
````

### 2.4 enumerate(): √≠ndice + valor

Cuando necesitas saber la posici√≥n de cada elemento:


````python
sucursales = ["Centro", "Tres R√≠os", "Plaza Fiesta", "Forum", "Galer√≠as"]

print("üìã Sucursales de Ver de Verdad en Culiac√°n:")
for i, sucursal in enumerate(sucursales, start=1):
    print(f"  {i}. {sucursal}")
````

### 2.5 List comprehension (listas por comprensi√≥n)

Una forma elegante de crear listas con un `for` en una sola l√≠nea:


````python
# Forma tradicional
cuadrados = []
for n in range(1, 11):
    cuadrados.append(n ** 2)
print(cuadrados)

# Forma con list comprehension ‚Äî una l√≠nea
cuadrados = [n ** 2 for n in range(1, 11)]
print(cuadrados)

# Con condici√≥n: solo n√∫meros pares
pares = [n for n in range(1, 21) if n % 2 == 0]
print(pares)

# Transformar una lista
nombres = ["hugo", "mar√≠a", "carlos", "ana"]
nombres_formateados = [nombre.title() for nombre in nombres]
print(nombres_formateados)
````

## 3. Ciclo while: repetir mientras se cumpla una condici√≥n

El `while` repite un bloque **mientras** la condici√≥n sea `True`. Es √∫til cuando no sabes cu√°ntas veces necesitas repetir.

### 3.1 while b√°sico


````python
# Contar del 1 al 5
contador = 1

while contador <= 5:
    print(f"Contando: {contador}")
    contador += 1  # ‚ö†Ô∏è Sin esto, el ciclo ser√≠a infinito

print("¬°Listo!")
````

### 3.2 while con input (men√∫ interactivo)


````python
# Calculadora interactiva
print("üßÆ Calculadora Simple")
print("Escribe 'salir' para terminar
")

while True:
    entrada = input("Escribe una operaci√≥n (ej: 5 + 3): ")

    if entrada.lower() == "salir":
        print("üëã ¬°Hasta luego!")
        break

    try:
        resultado = eval(entrada)  # eval() eval√∫a la expresi√≥n
        print(f"  = {resultado}
")
    except:
        print("  ‚ö†Ô∏è Operaci√≥n no v√°lida, intenta de nuevo
")
````

### 3.3 while con acumulador


````python
# Sumar hasta llegar a un objetivo
meta = 50000
ventas_totales = 0
dia = 0

print(f"üéØ Meta de ventas: ${meta:,}
")

while ventas_totales < meta:
    dia += 1
    # Simulamos ventas aleatorias entre 5000 y 15000
    import random
    venta_dia = random.randint(5000, 15000)
    ventas_totales += venta_dia
    print(f"D√≠a {dia}: ${venta_dia:,} ‚Äî Total: ${ventas_totales:,}")

print(f"
‚úÖ ¬°Meta alcanzada en {dia} d√≠as!")
print(f"Total vendido: ${ventas_totales:,}")
````

## 4. Control de flujo: break, continue, pass

### 4.1 break ‚Äî salir del ciclo inmediatamente


````python
# Buscar un producto en inventario
inventario = ["lentes", "armaz√≥n", "soluci√≥n", "estuche", "microfibra", "gotas"]
buscar = "estuche"

for producto in inventario:
    print(f"Revisando: {producto}...")
    if producto == buscar:
        print(f"‚úÖ ¬°Encontrado: {buscar}!")
        break  # Ya lo encontr√©, no necesito seguir buscando
else:
    # El else del for se ejecuta si NO se hizo break
    print(f"‚ùå {buscar} no est√° en inventario")
````

### 4.2 continue ‚Äî saltar a la siguiente iteraci√≥n


````python
# Procesar solo n√∫meros positivos
numeros = [10, -3, 25, -7, 0, 15, -1, 30]

print("Procesando solo n√∫meros positivos:")
suma = 0
for n in numeros:
    if n <= 0:
        continue  # Salta este n√∫mero y pasa al siguiente
    suma += n
    print(f"  + {n} = {suma}")

print(f"
Suma de positivos: {suma}")
````

## 5. Ciclos anidados

Un ciclo dentro de otro ciclo. √ötil para tablas, matrices y combinaciones.


````python
# Tablas de multiplicar del 1 al 5
for tabla in range(1, 6):
    print(f"
--- Tabla del {tabla} ---")
    for multiplicador in range(1, 11):
        print(f"  {tabla} √ó {multiplicador} = {tabla * multiplicador}")
````

````python
# Patr√≥n de estrellas (tri√°ngulo)
filas = 5
for i in range(1, filas + 1):
    print("‚≠ê" * i)
````

````python
# Combinaciones de productos
armazones = ["Negro", "Caf√©", "Azul"]
lentes = ["Monofocal", "Bifocal", "Progresivo"]

print("üìã Combinaciones disponibles:")
contador = 0
for armazon in armazones:
    for lente in lentes:
        contador += 1
        print(f"  {contador}. Armaz√≥n {armazon} + Lente {lente}")

print(f"
Total de combinaciones: {contador}")
````

### Ejercicio 2: Validaci√≥n de contrase√±a


````python
# Validar que una contrase√±a cumpla ciertos requisitos
password = input("Crea una contrase√±a: ")

tiene_mayuscula = False
tiene_minuscula = False
tiene_numero = False
tiene_largo = len(password) >= 8

for caracter in password:
    if caracter.isupper():
        tiene_mayuscula = True
    elif caracter.islower():
        tiene_minuscula = True
    elif caracter.isdigit():
        tiene_numero = True

print(f"
üìã Revisi√≥n de contrase√±a:")
print(f"  {'‚úÖ' if tiene_largo else '‚ùå'} M√≠nimo 8 caracteres ({len(password)} caracteres)")
print(f"  {'‚úÖ' if tiene_mayuscula else '‚ùå'} Al menos una may√∫scula")
print(f"  {'‚úÖ' if tiene_minuscula else '‚ùå'} Al menos una min√∫scula")
print(f"  {'‚úÖ' if tiene_numero else '‚ùå'} Al menos un n√∫mero")

if all([tiene_largo, tiene_mayuscula, tiene_minuscula, tiene_numero]):
    print("
üîí ¬°Contrase√±a segura!")
else:
    print("
‚ö†Ô∏è La contrase√±a no cumple todos los requisitos")
````

### Ejercicio 3: N√∫mero secreto


````python
# Juego de adivina el n√∫mero
import random

numero_secreto = random.randint(1, 100)
intentos = 0
max_intentos = 7

print("üéÆ ¬°Adivina el n√∫mero!")
print(f"Estoy pensando en un n√∫mero del 1 al 100")
print(f"Tienes {max_intentos} intentos
")

while intentos < max_intentos:
    intentos += 1
    intento_restante = max_intentos - intentos

    respuesta = int(input(f"Intento {intentos}/{max_intentos}: "))

    if respuesta == numero_secreto:
        print(f"
üéâ ¬°GANASTE! El n√∫mero era {numero_secreto}")
        print(f"Lo adivinaste en {intentos} intentos")
        break
    elif respuesta < numero_secreto:
        print(f"  ‚¨ÜÔ∏è M√°s alto... (quedan {intento_restante})")
    else:
        print(f"  ‚¨áÔ∏è M√°s bajo... (quedan {intento_restante})")
else:
    print(f"
üòî ¬°Se acabaron los intentos! El n√∫mero era {numero_secreto}")
````

## 7. üèÜ Mini Proyecto: Caja registradora

Vamos a crear una caja registradora que combine todo lo aprendido:


````python
# üèÜ Mini Proyecto: Caja Registradora de √ìptica

productos = {
    "1": ("Consulta visual", 99),
    "2": ("Lentes monofocales", 890),
    "3": ("Lentes bifocales", 1490),
    "4": ("Lentes progresivos", 2490),
    "5": ("Armaz√≥n b√°sico", 350),
    "6": ("Armaz√≥n premium", 890),
    "7": ("Soluci√≥n limpiadora", 120),
    "8": ("Estuche", 80),
}

carrito = []

print("=" * 50)
print("  üè™ √ìPTICA VER DE VERDAD ‚Äî Caja Registradora")
print("=" * 50)
print("
üìã Productos disponibles:")
for codigo, (nombre, precio) in productos.items():
    print(f"  [{codigo}] {nombre:.<30} ${precio:,.0f}")

print("
Escribe el n√∫mero del producto para agregarlo")
print("Escribe 'cobrar' para terminar
")

while True:
    opcion = input("Producto: ").strip().lower()

    if opcion == "cobrar":
        break

    if opcion in productos:
        nombre, precio = productos[opcion]
        carrito.append((nombre, precio))
        print(f"  ‚úÖ {nombre} (${precio:,.0f}) agregado al carrito")
    else:
        print("  ‚ö†Ô∏è Opci√≥n no v√°lida")

# Generar ticket
if len(carrito) == 0:
    print("
üõí El carrito est√° vac√≠o")
else:
    subtotal = sum(precio for _, precio in carrito)
    descuento = 0

    # Descuento por monto
    if subtotal >= 3000:
        descuento = subtotal * 0.10
        tipo_descuento = "10% por compra mayor a $3,000"
    elif subtotal >= 1500:
        descuento = subtotal * 0.05
        tipo_descuento = "5% por compra mayor a $1,500"

    total = subtotal - descuento

    print()
    print("=" * 50)
    print("  üßæ TICKET DE COMPRA")
    print("  √ìptica Ver de Verdad")
    print("=" * 50)

    for nombre, precio in carrito:
        print(f"  {nombre:.<35} ${precio:>8,.0f}")

    print("-" * 50)
    print(f"  {'Subtotal':.<35} ${subtotal:>8,.0f}")

    if descuento > 0:
        print(f"  {'Descuento (' + tipo_descuento + ')':.<35}")
        print(f"  {'':.<35} -${descuento:>7,.0f}")

    print("=" * 50)
    print(f"  {'TOTAL':.<35} ${total:>8,.0f}")
    print("=" * 50)
    print(f"
  Art√≠culos: {len(carrito)}")
    print("  ¬°Gracias por su compra! üëì")
````
