## 5. Python como calculadora üßÆ

Python puede hacer operaciones matem√°ticas. No necesitas `print()` en Colab ‚Äî la √∫ltima l√≠nea de una celda se muestra autom√°ticamente.

### Operaciones adicionales

| Operaci√≥n | S√≠mbolo | Ejemplo | Resultado |
|-----------|---------|---------|-----------|
| Suma | `+` | `10 + 5` | `15` |
| Resta | `-` | `100 - 37` | `63` |
| Multiplicaci√≥n | `*` | `12 * 8` | `96` |
| Divisi√≥n | `/` | `144 / 12` | `12.0` |
| Potencia | `**` | `2 ** 10` | `1024` |
| Divisi√≥n entera | `//` | `17 // 5` | `3` |
| Residuo (m√≥dulo) | `%` | `17 % 5` | `2` |

## 6. Variables: guardar informaci√≥n üì¶

Una **variable** es como una caja con nombre donde guardas un valor. As√≠ no tienes que repetir datos.


````python
# Creamos variables (se usa = para asignar)
nombre = "Culiac√°n"
poblacion = 1003530
es_capital = True

# Las usamos
print(nombre)
print(poblacion)
print(es_capital)
````

### Reglas para nombres de variables

‚úÖ **V√°lido:** `mi_nombre`, `edad2`, `total_ventas`, `ciudad_1`  
‚ùå **No v√°lido:** `2edad` (no empieza con n√∫mero), `mi nombre` (no lleva espacios), `print` (no uses palabras reservadas)

> üí° **Tip:** Usa nombres descriptivos. `tv` no dice nada, pero `total_ventas` s√≠.


````python
# Las variables se pueden usar en operaciones
precio = 250
cantidad = 3
total = precio * cantidad

print("El total es:", total)
````

````python
# Tambi√©n se pueden actualizar
contador = 0
print("Antes:", contador)

contador = contador + 1
print("Despu√©s:", contador)
````

## 7. Tipos de datos b√°sicos

Python reconoce autom√°ticamente qu√© tipo de dato le est√°s dando:


````python
# Texto (string / str) ‚Äî siempre entre comillas
ciudad = "Culiac√°n"
print(type(ciudad))

# N√∫mero entero (int)
edad = 25
print(type(edad))

# N√∫mero decimal (float)
temperatura = 38.5
print(type(temperatura))

# Verdadero o falso (bool)
es_viernes = True
print(type(es_viernes))
````

| Tipo | En Python | Ejemplo |
|------|-----------|---------|
| Texto | `str` | `"Hola"`, `'Python'` |
| N√∫mero entero | `int` | `42`, `-7`, `1000` |
| N√∫mero decimal | `float` | `3.14`, `99.99` |
| Verdadero/Falso | `bool` | `True`, `False` |

## 8. Jugando con texto (strings) ‚úèÔ∏è

Los strings se pueden combinar y manipular:


````python
# Unir texto (concatenar)
saludo = "Hola"
nombre = "Culiac√°n"
mensaje = saludo + " " + nombre + "!"
print(mensaje)
````

````python
# f-strings: la forma moderna y m√°s f√°cil de combinar texto con variables
nombre = "Hugo"
ciudad = "Culiac√°n"
print(f"Me llamo {nombre} y soy de {ciudad}")
````

````python
# Algunos trucos con strings
texto = "culiac√°n inteligencia artificial"

print(texto.upper())          # Todo en may√∫sculas
print(texto.title())          # Primera letra de cada palabra en may√∫scula
print(texto.replace("artificial", "aumentada"))  # Reemplazar texto
print(len(texto))             # Longitud del texto (n√∫mero de caracteres)
````
